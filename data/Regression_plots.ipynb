```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dnn = pd.read_csv("C:/workspaces/project/kaggle/result/reg_dnn.csv")
df_ols = pd.read_csv("C:/workspaces/project/kaggle/result/reg_ols.csv")
df_rf = pd.read_csv("C:/workspaces/project/kaggle/result/reg_rf.csv")
df_xgb = pd.read_csv("C:/workspaces/project/kaggle/result/reg_xgboost.csv")
df_svm = pd.read_csv("C:/workspaces/project/kaggle/result/reg_svm.csv")
```

```python
x1 = df_dnn['y_true']
y1 = df_dnn['pred']
```

```python
x2 = df_ols['y_true']
y2 = df_ols['pred']
```

```python
x3 = df_rf['Y_true']
y3 = df_rf['Y_pred']
```

```python
x4 = df_xgb['Y_true']
y4 = df_xgb['Y_pred']
```

```python
x5 = df_svm['Y_true']
y5 = df_svm['Y_pred']
```

##### Random Forest 시각화

```python
plt.figure(figsize=(8,8))
#plt.rc('font', size=10)
plt.rc('axes', titlesize=25)   
plt.rc('axes', labelsize=20)  
plt.rc('xtick', labelsize=15)   
plt.rc('ytick', labelsize=15) 
plt.rc('legend', fontsize=13)
#plt.rc('figure', titlesize=20)

plt.plot(x3, y3, 'x', linewidth=4.5)
plt.plot([0, 100], [0, 100], 'k--', linewidth=2, label="Random_Forest")
plt.ylim(0,20)
plt.xlim(0,80)
plt.xlabel('True')
plt.ylabel('Predicted')
plt.title('Linear Regression - Random_Forest')
plt.legend()

plt.savefig('result/images/reg_rf.png', dpi=300)
```

##### Xgboost 시각화

```python
plt.figure(figsize=(8,8))
#plt.rc('font', size=10)
plt.rc('axes', titlesize=25)   
plt.rc('axes', labelsize=20)  
plt.rc('xtick', labelsize=15)   
plt.rc('ytick', labelsize=15) 
plt.rc('legend', fontsize=13)
#plt.rc('figure', titlesize=20)

plt.plot(x4, y4, 'x', linewidth=4.5)
plt.plot([0, 100], [0, 100], 'k--', linewidth=2, label="XGBoost")
plt.ylim(0,20)
plt.xlim(0,80)
plt.xlabel('True')
plt.ylabel('Predicted')
plt.title('Linear Regression - XGBoost')
plt.legend()

plt.savefig('result/images/reg_xgb.png', dpi=300)
```

##### 합쳐서 시각화

```python
plt.figure(figsize=(8,8))
#plt.rc('font', size=10)
plt.rc('axes', titlesize=25)   
plt.rc('axes', labelsize=20)  
plt.rc('xtick', labelsize=15)   
plt.rc('ytick', labelsize=15) 
plt.rc('legend', fontsize=13)
#plt.rc('figure', titlesize=20)

plt.plot(x1, y1, 'x', linewidth=4.5, label = "DNN")
plt.plot(x2, y2, 'x', linewidth=4.5, label = "OLS")
plt.plot(x3, y3, 'x', linewidth=4.5, label = "Random_Forest")
plt.plot(x4, y4, 'x', linewidth=4.5, label = "XGBoost")
plt.plot(x5, y5, 'x', linewidth=4.5, label = "SVM")
plt.plot([0, 100], [0, 100], 'k--', linewidth=2, label="reg_line")
plt.ylim(0,40)
plt.xlim(0,80)
plt.xlabel('True Value')
plt.ylabel('Predicted Value')
plt.title('Linear Regression')
plt.legend()

plt.savefig('result/images/reg.png', dpi=300)
```

##### 시각화

```python
plt.figure(figsize=(12,12))
#plt.rc('font', size=10)
plt.rc('axes', titlesize=20)   
plt.rc('axes', labelsize=15)  
plt.rc('xtick', labelsize=15)   
plt.rc('ytick', labelsize=15) 
plt.rc('legend', fontsize=12)
#plt.rc('figure', titlesize=20)

plt.subplot(2,3,1)
plt.tight_layout(pad=8.0)
plt.plot(x1, y1, 'x', color='b',linewidth=3, label = "DNN")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Linear Regression - DNN')
plt.legend()

plt.subplot(2,3,2)
plt.plot(x2, y2, 'x', color='orange',linewidth=3, label = "OLS")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Linear Regression - OLS')
plt.legend()

plt.subplot(2,3,3)
plt.plot(x3, y3, 'x', color='g',linewidth=3, label = "Random_Forest")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - rand_forest')
plt.legend()

plt.subplot(2,3,4)
plt.plot(x4, y4, 'x', color='r',linewidth=3, label = "XGBoost")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Linear Regression - XGBoost')
plt.legend()

plt.subplot(2,3,5)
plt.plot(x5, y5, 'x', color='k',linewidth=3, label = "SVM")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Linear Regression - SVM')
plt.legend()
plt.show()

plt.savefig('result/images/reg_sub.png', dpi=450)
```
