```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_dnn = pd.read_csv("C:/workspaces/project/kaggle/result/clf_dnn.csv")
df_skl = pd.read_csv("C:/workspaces/project/kaggle/result/clf_skl.csv")
df_rf = pd.read_csv("C:/workspaces/project/kaggle/result/clf_rf.csv")
df_xgb = pd.read_csv("C:/workspaces/project/kaggle/result/clf_xgboost.csv")
# df_svm = pd.read_csv("C:/workspaces/project/kaggle/result/clf_svm.csv")
```

```python
x1dnn = df_dnn['fpr']
y1dnn = df_dnn['tpr']
```

```python
x2skl = df_skl['fpr']
y2skl = df_skl['tpr']
```

```python
x3rf = df_rf['fpr']
y3rf = df_rf['tpr']
```

```python
x4xgb = df_xgb['fpr']
y4xgb = df_xgb['tpr']
```

```python
# x5svm = df_svm['fpr']
# y5svm = df_svm['tpr']
```

##### Random Forest 시각화

```python
plt.figure(figsize=(8,8))
#plt.rc('font', size=10)
plt.rc('axes', titlesize=25)   
plt.rc('axes', labelsize=20)  
plt.rc('xtick', labelsize=15)   
plt.rc('ytick', labelsize=15) 
plt.rc('legend', fontsize=13)
#plt.rc('figure', titlesize=20)

plt.plot(x3rf, y3rf, '-.', linewidth=4.5, label = "Random_Forest")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - Random_Forest')
plt.legend()

plt.savefig('result/images/roc_rf.png', dpi=300)
```

##### Xgboost 시각화

```python
plt.figure(figsize=(8,8))
#plt.rc('font', size=10)
plt.rc('axes', titlesize=25)   
plt.rc('axes', labelsize=20)  
plt.rc('xtick', labelsize=15)   
plt.rc('ytick', labelsize=15) 
plt.rc('legend', fontsize=13)
#plt.rc('figure', titlesize=20)

plt.plot(x4xgb, y4xgb, '-.', linewidth=4.5, label = "XGBoost")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - XGBoost')
plt.legend()

plt.savefig('result/images/roc_xgb.png', dpi=300)
```

##### 합쳐서 시각화

```python
plt.figure(figsize=(8,8))
#plt.rc('font', size=10)
plt.rc('axes', titlesize=25)   
plt.rc('axes', labelsize=20)  
plt.rc('xtick', labelsize=15)   
plt.rc('ytick', labelsize=15) 
plt.rc('legend', fontsize=13)
#plt.rc('figure', titlesize=20)

plt.plot(x1dnn, y1dnn, ':', linewidth=4.5, label = "DNN")
plt.plot(x2skl, y2skl, '--', linewidth=4.5, label = "skl")
plt.plot(x3rf, y3rf, '-.', linewidth=4.5, label = "Random_Forest")
plt.plot(x4xgb, y4xgb, linewidth=4.5, label = "XGBoost")
# plt.plot(x5svm, y5svm, linewidth=4.5, label = "SVM")
plt.plot([0, 1], [0, 1], 'k--', linewidth=2, label="random guess")
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification')
plt.legend()

plt.savefig('result/images/roc.png', dpi=300)
```

##### 시각화

```python
plt.figure(figsize=(12,12))
#plt.rc('font', size=10)
plt.rc('axes', titlesize=20)   
plt.rc('axes', labelsize=15)  
plt.rc('xtick', labelsize=15)   
plt.rc('ytick', labelsize=15) 
plt.rc('legend', fontsize=12)
#plt.rc('figure', titlesize=20)

plt.subplot(2,3,1)
plt.tight_layout(pad=8.0)
plt.plot(x1dnn, y1dnn, ':', color='b',linewidth=4.5, label = "DNN")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - DNN')
plt.legend()

plt.subplot(2,3,2)
plt.plot(x2skl, y2skl, '--', color='orange',linewidth=4.5, label = "skl")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - skl')
plt.legend()

plt.subplot(2,3,3)
plt.plot(x3rf, y3rf, '-.', color='g',linewidth=4.5, label = "Random_Forest")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - Random_Forest')
plt.legend()

plt.subplot(2,3,4)
plt.plot(x4xgb, y4xgb, color='r',linewidth=4.5, label = "XGBoost")
plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Classification - XGBoost')
plt.legend()

# plt.subplot(2,3,5)
# plt.plot(x5svm, y5svm, color='k',linewidth=4.5, label = "SVM")
# plt.plot([0, 1], [0, 1], 'k--',linewidth=2)
# plt.xlabel('False positive rate')
# plt.ylabel('True positive rate')
# plt.title('Classification - SVM')
# plt.legend()
plt.show()

plt.savefig('result/images/roc_sub.png', dpi=300)
```
